### First, I need to find the mappings from tickers to CIK 

```python
import pandas as pd
companies = pd.read_csv("../finance/biotech-companylist.csv")
companies
```

```python
import re
import urllib2
     
def get_cik(ticker):
    ''' this function uses yahoo to translate a ticker into a CIK '''
    url = "http://finance.yahoo.com/q/sec?s=%s+SEC+Filings" % (ticker)
    try:
        tick = int(re.findall('=[0-9]*', str(re.findall('cik=[0-9]*', urllib2.urlopen(url).read())[0]))[0][1:])
    except:
        tick = -1
    return tick
```

```python
symbols = companies["Symbol"]
for s in range(10):
    print symbols[s] + ":" + str(get_cik(symbols[s]))
```

```python
# A dictionary holds the mappings from tickers to CIK
import time
import random
tick_to_CIK = {}
for i in range(len(symbols)):
    cik = get_cik(symbols[i])
    if cik != -1:
        tick_to_CIK[symbols[i]] = cik
    print symbols[i] + ":" + str(cik)
    interval = random.randint(2, 10)
    print str(i) + ": sleep for "  + str(interval) + " secs...."
    #time.sleep(interval)
```

```python
# Write the dictionary holding tick to CIK# mappings to a file
def writeDict(dict, filename):
    with open(filename, "a") as f:
        for i in dict.keys():            
            f.write(i + ":" + str(dict[i]) + "\n")

# Read in the dictionary holding tick to CIK# mappings from a file
def readDict(filename):
    with open(filename, "r") as f:
        dict = {}
        for line in f:
            values = line.split(":")
        dict[values[0]] = int(values[1].strip())
    return(dict)
```

```python
# Parse the lines in a crawler.idx file from SEC and return the URL pointing to
# the file
import re
def parse_crawler_idx(line):

    line = line.strip()
    company = line[0:61].strip()
    formType = line[62:73].strip()
    cik = line[74:85].strip()
    date_filed = line[86:97].strip()
    url_link = line[98:178].strip()
    if re.match(r'\d\d\d\d-\d\d-\d\d', date_filed):
        return company, formType, cik, date_filed, url_link

    return None, None, None, None, None
```
